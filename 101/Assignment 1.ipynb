```python
import pandas as pd
import numpy as np
```

```python
#import of the csv file
df = pd.read_csv (r'heart.csv')

df = df.fillna(df.median())

#change of the datatype to only one form eariler trestbps was of datatype float
df = df.astype({'trestbps':int ,'chol':int ,'oldpeak':int})
df
```

```python
#drop of all the duplicate values in the database which is stored in variable df
df = df.drop_duplicates()
```

```python
x = df.iloc [ : , :13]
#setting the values of x
```

```python
y = df.iloc [ : , 13: ]
#setting the values of y [( :  )->for row ( : )-> for columns]
```

```python
from sklearn.model_selection import train_test_split
#import of train test and split library
```

```python
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size = 0.25)
                                                # here the size is made of the ration 75% - 25%
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
#import the model
```

```python
reg = LogisticRegression()
#declaring the module for furthur use
```

```python
reg.fit(x_train,y_train)
# getting the meadian value for evaluaton
```

```python
yPrediction = reg.predict(x_test)
#calculating prediction we use test value to calculate prediction and cause ytest for tesing purpose only 
#we are using xtest here
```

```python
#from here onward their is all import of libraries and using them
from sklearn.metrics import classification_report
```

```python
yPrediction.shape
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,yPrediction))
```

```python
print(classification_report(y_test,yPrediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,yPrediction))
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(y_test,yPrediction),annot = True)
```
